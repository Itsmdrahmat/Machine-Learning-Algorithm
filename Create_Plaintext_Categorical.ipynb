```python
import nltk
nltk.download('brown')
from nltk.corpus import brown
brown.categories()
```

```python

from nltk.corpus.reader import CategorizedPlaintextCorpusReader
reader = CategorizedPlaintextCorpusReader('.', r'movie_.*\.txt', cat_pattern = r'movie_(\w+)\.txt')
print ("Categorize : ", reader.categories())
print ("\nNegative field : ", reader.fileids(categories =['neg']))
print ("\nPositive field : ", reader.fileids(categories =['pos']))
from nltk.corpus.reader import CategorizedPlaintextCorpusReader
 
reader = CategorizedPlaintextCorpusReader('.', r'movie_.*\.txt', cat_map ={'movie_pos.txt': ['pos'],'movie_neg.txt': ['neg']})    
print ("Categorize : ", reader.categories())
```
