# **Text Pre-processing using NLTK**

# Stemming

```python
from nltk.stem import PorterStemmer as ps
text1=("The Students of MSc IT are using historical data for data for data warehouse projects.").split()
print(text1)
for w in text1:
  rootWord = ps().stem(w)
  print(rootWord)
```

```python
from nltk.stem import PorterStemmer as ps
text2=["Unexpected", "Disagreement", "Disagree", "Agreement", "Quirkiness", "Historical", "Canonical", "Unkind"]
print(text2)
for w in text2:
  stemPrint = ps().stem(w)
  print(w, " -Stem-", stemPrint)
```
