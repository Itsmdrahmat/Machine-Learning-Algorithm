**4b. RegEx**

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import RegexpTokenizer
```

```python
tk = RegexpTokenizer('\s+', gaps= True)
```

```python
str = "The purpose of a grammar is to give an explicit description of a language"
```

```python
tokens = tk.tokenize(str)
print (tokens)
```

**4c. Tokenization using nltk**
