## **TreeBankTokenizer**

```python
import nltk
nltk.__file__
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.tokenize import TreebankWordTokenizer
from nltk.tokenize import RegexpTokenizer
from nltk.tokenize import WordPunctTokenizer
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize("Hello World this is TreebankWordTokenizer")
```

```python
tokenizer = WordPunctTokenizer()
tokenizer.tokenize("Hello World this is WordPunctTokenizer")
```

```python
tokenizer = RegexpTokenizer("[\w']+")
tokenizer.tokenize("Hello World's this is RegexpTokenizer")
```

```python
from nltk import tokenize
tokenizer = RegexpTokenizer("[\w']+")
tokenizer.tokenize("abc@gmail.com")
```

```python
tokenizer.tokenize("xyz@rediffmail.com")
```

```python

tokenizer.tokenize("rjc@rjcollege.edu.in")
```

## **Steming**

```python
text=("This is Stemming Example1").split()
print(text)
for w in text:
  rootWord=PorterStemmer().stem(w)
  print(rootWord)
```

```python
words=["Unexpected", "disagreement", "historical"]
for w in words:
  rootWord=PorterStemmer().stem(w)
  print(w,"-Stem-", rootWord)
```

## **Lemmatization**

```python
wordnet_lemmatizer = WordNetLemmatizer()
lemtext = "This is example of Lemmation"
tokenization = nltk.word_tokenize(lemtext)
for w in tokenization:
  print("Lemma for {} is {}".format(w , wordnet_lemmatizer.lemmatize(w)))
```

```python
from nltk.corpus import gutenberg
gutenberglemm = gutenberg.raw('austen-emma.txt')
wordnet_lemmatizer = WordNetLemmatizer()
tokenization = nltk.word_tokenize(gutenberglemm)
for w in tokenization:
  print("Lemma for {} is {}".format(w , wordnet_lemmatizer.lemmatize(w)))
```
